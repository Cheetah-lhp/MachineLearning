```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import accuracy_score
# Set seed for reproducibility
np.random.seed(0)

# Create 100 samples and 10 features
n_samples, n_features = 100, 10
X = np.random.randn(n_samples, n_features)

# Only the first 3 features are important
true_coefs = np.array([5, -3, 1.5] + [0]*(n_features - 3))
y = X @ true_coefs + np.random.randn(n_samples) * 0.5  # add noise #X @ true_coefs: This part performs a matrix multiplication between the features (X) and the true coefficients (true_coefs).

# Fit models
lr = LinearRegression().fit(X, y)
ridge = Ridge(alpha=0.01).fit(X, y)
lasso = Lasso(alpha=0.01).fit(X, y)

# Plot coefficients
plt.figure(figsize=(10, 6))
plt.plot(true_coefs, label='True coefficients', linestyle='--', marker='o')
plt.plot(lr.coef_, label='Linear Regression', marker='x')
plt.plot(ridge.coef_, label='Ridge Regression', marker='s')
plt.plot(lasso.coef_, label='Lasso Regression', marker='^')
plt.legend()
plt.title("Comparison of Coefficients")
plt.xlabel("Feature Index")
plt.ylabel("Coefficient Value")
plt.grid(True)
plt.show()
```

```python
# Tạo tập dữ liệu test mới
X_test_large = np.random.randn(100, n_features)
y_test_large = X_test_large @ true_coefs + np.random.randn(100) * 0.5  # thêm noise

# Dự đoán
y_pred_lr = lr.predict(X_test_large)
y_pred_ridge = ridge.predict(X_test_large)
y_pred_lasso = lasso.predict(X_test_large)

print("accuracy = ", lr.score(X_test_large, y_test_large))
print("accuracy = ", ridge.score(X_test_large, y_test_large))
print("accuracy = ", lasso.score(X_test_large, y_test_large))
```
