```python
import numpy as np
np.random.seed(10)
def f(x):
  return 2*x + 1

X = np.random.randn(20)
Y = f(X) + np.random.randn(20) * 0.5

print(X)
print(Y)
```

```python
import matplotlib.pyplot as plt
plt.scatter([X], Y)
plt.show()
```

```python
from sklearn import datasets, linear_model

# fit the model by Linear Regression
regr = linear_model.LinearRegression()
regr.fit(X.reshape(-1, 1), Y.reshape(-1, 1))

regr_lasso = linear_model.Lasso(alpha=0.)
regr_lasso.fit(X.reshape(-1, 1), Y.reshape(-1, 1))

regr_ridge = linear_model.Ridge(alpha=1)
regr_ridge.fit(X.reshape(-1, 1), Y.reshape(-1, 1))

# Compare two results
print('Linear Regression model: ', regr.coef_, regr.intercept_ )
print('Lasso model: ', regr_lasso.coef_, regr_lasso.intercept_)
print('Ridge model: ', regr_ridge.coef_, regr_ridge.intercept_)
```

```python
plt.scatter([X], Y)
x_draw = np.array([min(X), max(X)])
y_draw = regr.coef_[0] * x_draw + regr.intercept_
y_draw_lasso = regr_lasso.coef_[0] * x_draw + regr_lasso.intercept_
y_draw_ridge = regr_ridge.coef_[0] * x_draw + regr_ridge.intercept_
print(x_draw)
print(y_draw)
plt.plot(x_draw, y_draw, 'r', label = 'Linear regression')
plt.plot(x_draw, y_draw_lasso, 'g', label = 'Lasso regression')
plt.plot(x_draw, y_draw_ridge, 'b', label = 'Ridge regression')
plt.legend()
plt.show()
```
