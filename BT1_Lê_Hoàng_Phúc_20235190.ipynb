### Đề bài: Huấn luyện mô hình Logistic Regression với bài toán AND

```python
import numpy as np
```

```python
# Khởi tạo dataset của bài toán AND:
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [0], [0], [1]])
```

```python
# Định nghĩa các hàm tính toán:
# Hàm sigmoid:
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Hàm loss:
def Loss(y_true, y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
# Khởi tạo các tham số:
m, n = X.shape
W = np.random.default_rng(29).random([n, 1])
b = 0
learning_rate = 0.1
epochs = 1000
```

```python
# Huấn luyện:
for epoch in range(epochs):
    z = np.dot(X, W) + b
    A = sigmoid(z)
    loss = Loss(Y, A)

    # Tính gradient
    dz = A - Y
    dW = np.dot(X.T, dz) / m
    db = np.sum(dz) / m

    # Cập nhật tham số
    W -= learning_rate * dW
    b -= learning_rate * db

    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch + 1}: Loss = {loss}")
```

```python
# Kết quả thu được:
print(f"Trọng số W: {W.flatten()}")
print(f"Bias b: {b}")
print()
print("Kết quả dự đoán:")
predictions = sigmoid(np.dot(X, W) + b)
for i in range(len(X)):
    res = 1 if predictions[i] >= 0.5 else 0
    print(f"Input: {X[i]} -> Xác suất: {predictions[i][0]:.5f} -> Nhãn: {res}")
```
