```python
import numpy as np
np.random.seed(10)
def f(x):
  return 2*x + 1

X = np.random.randn(20)
Y = f(X) + np.random.randn(20) * 0.5

print(X)
print(Y)
```

```python
import matplotlib.pyplot as plt
plt.scatter([X], Y)
plt.show()
```

```python
from sklearn import datasets, linear_model

# fit the model by Linear Regression
regr = linear_model.Ridge(fit_intercept=True, alpha = 10)
regr.fit(X.reshape(-1, 1), Y.reshape(-1, 1))

# Compare two results
print( 'Solution found by scikit-learn  : ', regr.coef_, regr.intercept_ )
```

```python
plt.scatter([X], Y)
x_draw = np.array([min(X), max(X)])
y_draw = regr.coef_[0] * x_draw + regr.intercept_
print(x_draw)
print(y_draw)
plt.plot(x_draw, y_draw, 'r')
plt.show()
```
